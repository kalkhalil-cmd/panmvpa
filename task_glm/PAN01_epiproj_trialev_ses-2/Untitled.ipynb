```python
from pathlib import Path

# 1. Define the directory path
target_dir = Path("/home/jovyan/panmvpa/task_glm/PAN01_epiproj_trialev_ses-1")

# 2. Look through all files in the directory
for file_path in target_dir.iterdir():
    
    # 3. Filter for files only (skip folders) and check conditions
    if file_path.is_file() and "beta" in file_path.name:
        
        # 4. Create the new filename
        new_name = file_path.name.replace("beta_map", "beta")
        new_file_path = file_path.with_name(new_name)
        
        # 5. Rename the file
        file_path.rename(new_file_path)
        print(f"Renamed: {file_path.name} -> {new_name}")
```
